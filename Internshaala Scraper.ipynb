```python
# Importing Libraries
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# Setting the driver for selenium
driver = webdriver.Chrome("C:/Users/malay/chromedriver_win32/chromedriver.exe")  
driver.maximize_window() 
a = driver.get("https://internshala.com/internships/matching-preferences")  
```

```python
# Getting the content of the page
content = driver.page_source
soup = BeautifulSoup(content, "html.parser")
```

```python
# Printing the content
print(soup.contents)
```

```python
# Seggregate id=content
results = soup.find(id = "content")
print(results.prettify())
```

```python
# Filter by id=internship_list_container_1
results1 = results.find(id = "internship_list_container_1")
print(results1.prettify())
```

```python
# Filter by Divisions for headings
results2 = results1.find_all("div", class_="heading_4_5 profile")
```

```python
# Filter by individual_internship_details
results3 = results1.find_all("div", class_="individual_internship_details individual_internship_job")
```

```python
# Get the list of all jobs for a specific query
for result in results3:
    locs = result.find_all("a")
    for loc in locs:
        if loc.text == "Bangalore" or loc.text == "bangalore":
            job_blr = [loc.parent.parent for loc in locs]
job_blr
```

```python
x = "https://internshaala.com"
for result in results2:
    links = result.find_all("a")
    for link in links:
        link_url = link["href"]
        print("Apply here:",x+link_url,end = "\n")
```
